Candidate-Elimination Algorithm


```python
#CandidateElimination
import numpy as np 
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/ML-LAB/Book2.csv')
concepts = np.array(data.iloc[:,0:-1])
print("\nInstances are:\n",concepts)
target = np.array(data.iloc[:,-1])
print("\nTarget Values are: ",target)

def learn(concepts, target): 
    specific_h = concepts[0].copy()
    print("\nInitialization of specific_h and genearal_h")
    print("\nSpecific Boundary: ", specific_h)
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print("\nGeneric Boundary: ",general_h)  

    for i, h in enumerate(concepts):
        print("\nInstance", i+1 , "is ", h)
        if target[i] == "yes":
            print("Instance is Positive ")
            for x in range(len(specific_h)): 
                if h[x]!= specific_h[x]:                    
                    specific_h[x] ='?'                     
                    general_h[x][x] ='?'
                   
        if target[i] == "no":            
            print("Instance is Negative ")
            for x in range(len(specific_h)): 
                if h[x]!= specific_h[x]:                    
                    general_h[x][x] = specific_h[x]                
                else:                    
                    general_h[x][x] = '?'        
        
        print("Specific Bundary after ", i+1, "Instance is ", specific_h)         
        print("Generic Boundary after ", i+1, "Instance is ", general_h)
        print("\n")

    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]    
    for i in indices:   
        general_h.remove(['?', '?', '?', '?', '?', '?']) 
    return specific_h, general_h 

s_final, g_final = learn(concepts, target)

print("Final Specific_h: ", s_final, sep="\n")
print("Final General_h: ", g_final, sep="\n")
```

```python
import numpy as np
from math import log
from scipy.stats import entropy
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split
#importing data set tennis
df=pd.read_csv("/content/drive/MyDrive/ML-LAB/Tennis2.csv")
df.drop(df.columns[0], axis=1, inplace=True)#removed Example Column from dataset
s,n=0,0
target = np.array(df)
print(target)
for i in target:
  if i[4]=='Yes':s+=1
  else:n+=1
# Entropy calculation for 1st column
sy,sn,oy,on,ry,rn=0,0,0,0,0,0
for i in target:
    if i[0]=='Sunny'      and i[4]=='Yes':sy+=1
    elif i[0]=='Sunny'    and i[4]=='No':sn+=1
    elif i[0]=='Rainy'    and i[4]=='Yes':ry+=1
    elif i[0]=='Rainy'    and i[4]=='No':rn+=1
    elif i[0]=='Overcast' and i[4]=='Yes':oy+=1
    elif i[0]=='Overcast' and i[4]=='No':on+=1
l1=np.array([sy,sn,ry,rn,oy,on])
print(l1)
l1 = [x for x in l1 if x != 0]
#print("after removing zeros list-1",l1) #after removing zeros

sunny=sy+sn
rain=ry+rn
overcast=oy+on
sunny_entropy=abs((sy/sunny)*log(sy/sunny,2))-((sn/sunny)*log(sn/sunny,2))
rainy_entropy=abs((ry/rain)*log(ry/rain,2))-((rn/rain)*log(rn/rain,2))
overcast_entropy=abs((oy/overcast)*log(oy/overcast,2))-((oy/overcast)*0)

print("sunny,rainy,overcast",sunny_entropy,rainy_entropy,overcast_entropy)

#Entropy for second column
hy,hn,my,mn,cy,cn=0,0,0,0,0,0
for i in (target):
    if   i[1]=='Hot'  and i[4]=='Yes':hy+=1
    elif i[1]=='Hot'  and i[4]=='No':hn+=1
    elif i[1]=='Mild' and i[4]=='Yes':my+=1
    elif i[1]=='Mild' and i[4]=='No':mn+=1
    elif i[1]=='Cool' and i[4]=='Yes':cy+=1
    elif i[1]=='Cool' and i[4]=='No':cn+=1
l2=np.array([hy,hn,my,mn,cy,cn])
#print(l2) #printing l2 values before removing zeros
l2 = [x for x in l2 if x != 0]
hot=hy+hn
mild=my+mn
cool=cy+cn
hot_entropy=abs((hy/hot)*log(hy/hot,2))-((hn/hot)*log(hn/hot,2))
mild_entropy=abs((my/mild)*log(my/mild,2))-((mn/mild)*log(mn/mild,2))
cool_entropy=abs((cy/cool)*log(cy/cool,2))-((cn/cool)*log(cn/cool,2))
print("Hot_Entropy, Mild_Entropy, Cool_Entropy",hot_entropy,mild_entropy,cool_entropy)

#Entropy for 3rd column
high_y,high_n,normal_y,normal_n=0,0,0,0
for i in target:
  if i[2]=='High' and i[4]=='Yes':high_y+=1
  elif i[2]=='High' and i[4]=='No':high_n+=1
  elif i[2]=='Normal' and i[4]=='Yes':normal_y+=1
  elif i[2]=='Normal' and i[4]=='No':normal_n+=1
l3=np.array([high_y,high_n,normal_y,normal_n])
l3 = [x for x in l3 if x != 0]
print(l3)
high=high_y+high_n
normal=normal_y+normal_n
high_entropy=abs((high_y/high)*log(high_y/high,2))-((high_n/high)*log(high_n/high,2))
normal_entropy=abs((normal_y/normal)*log(normal_y/normal,2))-((normal_n/normal)*log(normal_n/normal,2))
print("High_Entropy,Normal_Entropy",high_entropy,normal_entropy)
#Entropy for 4th column
s_y,s_n,w_y,w_n=0,0,0,0
for i in target:
  if i[3]=='Strong' and i[4]=='Yes':s_y+=1
  elif i[3]=='Strong' and i[4]=='No':s_n+=1
  elif i[3]=='Weak' and i[4]=='Yes':w_y+=1
  elif i[3]=='Weak' and i[4]=='No':w_n+=1
l4=np.array([s_y,s_n,w_y,w_n])
l4 = [x for x in l4 if x != 0]  #removing zeroes if exists
strong=s_y+s_n
weak=w_y+w_n
strong_entropy=abs((s_y/strong)*log(s_y/strong,2))-((s_n/strong)*log(s_n/strong,2))
weak_entropy=abs((w_y/weak)*log(w_y/weak,2))-((w_n/weak)*log(w_n/weak,2))
print("Strong_Entropy,Weak_Entropy",strong_entropy,weak_entropy)
total_entropy=abs(-(s/len(df)*log(s/len(df),2))-(n/len(df)*log(n/len(df),2)))
print("Total_Entropy",total_entropy)


#Information Gain for each feature
outlook=((sunny/len(df)*sunny_entropy)+(rain/len(df)*rainy_entropy)+(overcast/len(df)*overcast_entropy))
temp=((hot/len(df)*hot_entropy)+(mild/len(df)*mild_entropy)+(cool/len(df)*cool_entropy))
humidity=((high/len(df)*high_entropy)+(normal/len(df)*normal_entropy))
wind=((strong/len(df)*strong_entropy)+(weak/len(df)*weak_entropy))
a=total_entropy-outlook
b=total_entropy-temp
c=total_entropy-humidity
d=total_entropy-wind
print("OutLook,Temperature,Humidity,Wind",a,b,c,d)




```

```python

#ID3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import copy

dataset = pd.read_csv('/content/Tennis2.csv')
X = dataset.iloc[:, 1:].values
# print(X)
attribute = ['outlook', 'temp', 'humidity', 'wind']


class Node(object):
    def __init__(self):
        self.value = None
        self.decision = None
        self.childs = None


def findEntropy(data, rows):
    yes = 0
    no = 0
    ans = -1
    idx = len(data[0]) - 1
    entropy = 0
    for i in rows:
        if data[i][idx] == 'Yes':
            yes = yes + 1
        else:
            no = no + 1

    x = yes/(yes+no)
    y = no/(yes+no)
    if x != 0 and y != 0:
        entropy = -1 * (x*math.log2(x) + y*math.log2(y))
    if x == 1:
        ans = 1
    if y == 1:
        ans = 0
    return entropy, ans


def findMaxGain(data, rows, columns):
    maxGain = 0
    retidx = -1
    entropy, ans = findEntropy(data, rows)
    if entropy == 0:
        """if ans == 1:
            print("Yes")
        else:
            print("No")"""
        return maxGain, retidx, ans

    for j in columns:
        mydict = {}
        idx = j
        for i in rows:
            key = data[i][idx]
            if key not in mydict:
                mydict[key] = 1
            else:
                mydict[key] = mydict[key] + 1
        gain = entropy

        
        for key in mydict:
            yes = 0
            no = 0
            for k in rows:
                if data[k][j] == key:
                    if data[k][-1] == 'Yes':
                        yes = yes + 1
                    else:
                        no = no + 1
            # print(yes, no)
            x = yes/(yes+no)
            y = no/(yes+no)
            # print(x, y)
            if x != 0 and y != 0:
                gain += (mydict[key] * (x*math.log2(x) + y*math.log2(y)))/14
        # print(gain)
        if gain > maxGain:
            # print("hello")
            maxGain = gain
            retidx = j
    #(maxGain,retidx,ans)                                                                                                                               
    return maxGain, retidx, ans


def buildTree(data, rows, columns):

    maxGain, idx, ans = findMaxGain(X, rows, columns)
    root = Node()
    root.childs = []
    # print(maxGain
    #
    # )
    if maxGain == 0:
        if ans == 1:
            root.value = 'Yes'
        else:
            root.value = 'No'
        return root

    root.value = attribute[idx]
    mydict = {}
    for i in rows:
        key = data[i][idx]
        if key not in mydict:
            mydict[key] = 1
        else:
            mydict[key] += 1

    newcolumns = copy.deepcopy(columns)
    newcolumns.remove(idx)
    for key in mydict:
        newrows = []
        for i in rows:
            if data[i][idx] == key:
                newrows.append(i)
        # print(newrows)
        temp = buildTree(data, newrows, newcolumns)
        temp.decision = key
        root.childs.append(temp)
    
    return root


def traverse(root):
    print(root.decision)
    print(root.value)
    l=[]
    
    n = len(root.childs)
    if n > 0:
        for i in range(0, n):
            
            traverse(root.childs[i])


def calculate():
    rows = [i for i in range(0, 14)]
    columns = [i for i in range(0, 4)]
    root = buildTree(X, rows, columns)
    root.decision = 'Start'
    traverse(root)

calculate()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.metrics import *
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
# Loading data
irisData = load_iris()
# Create feature and target arrays
X = irisData.data
y = irisData.target
# Split into training and test set
X_train, X_test,y_train,y_test = map(np.array,train_test_split(X,y, test_size=0.20))
print(f'size of training data {X_train.shape}')
print(f'size of testing data {X_test.shape}')
print('label 0: Setosa')
print('label 1: Versicolor')
print('label 2: Virginica')

classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print('results of classification with k=1')
for i in range(len(X_test)):
    print(f'sample: {X_test[i]}\tpredicted-label: {classifier.predict([X_test[i]])[0]} \t Actual-labelio : {y_test[i]}')

print('confusion matrix')
print(confusion_matrix(y_test, y_pred))

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
